```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6) # Set figure size
print("Libraries imported successfully! Starting to load data...")
```

```python
try:
    df = pd.read_csv('TSLA.csv')
    print("Data loaded successfully!")
except FileNotFoundError:
    print("Error: File not found. Please check if the filename is 'TSLA.csv' and it is in the same folder.")
```

```python
print("--- Raw Data Preview ---")
display(df.head())
print("n--- Missing Values Check ---")
print(df.isnull().sum())
if 'date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    print("nData cleaning complete: Date format converted.")
else:
    print("nWarning: 'Date' column not found.")
```

```python
#Calculate Daily Return pct_change() calculates percentage change, i.e., (Today's Price - Yesterday's Price) / Yesterday's Price
df['Daily Return'] = df['Close'].pct_change()
#calculate Moving Average
#Calculate the average closing price of the past 20 days, representing short-term trends
df['MA_20'] = df['Close'].rolling(window=20).mean()
#Calculate the average closing price of the past 50 days, representing medium-term trends
df['MA_50'] = df['Close'].rolling(window=50).mean()
print("Analysis metrics calculated: Daily Return, 20-Day MA, 50-Day MA.")
print("n--- Processed Data Preview (Check the last few columns) ---")
display(df.tail())
```

```python
#Chart 1: Tesla Closing Price vs Moving Averages
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Closing Price', alpha=0.5, color='gray')
plt.plot(df['Date'], df['MA_20'], label='20-Day Moving Average', color='blue')
plt.plot(df['Date'], df['MA_50'], label='50-Day Moving Average', color='red')
plt.title('Tesla (TSLA) Stock Price & Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()

#Chart 2: Daily Return Distribution
plt.figure(figsize=(14, 5))
sns.histplot(df['Daily Return'].dropna(), bins=50, color='green', kde=True)
plt.title('Distribution of Tesla Daily Returns')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')
plt.show()
```

```python
# Summary Statistics
print("Data Analysis Summary")
print("-" * 30)

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
else:
    df.index = pd.to_datetime(df.index)
# -----------------------------

# Find dates with max gain and max loss
max_gain = df.loc[df['Daily Return'].idxmax()]
max_loss = df.loc[df['Daily Return'].idxmin()]

date_gain = max_gain.name.strftime('%Y-%m-%d')
date_loss = max_loss.name.strftime('%Y-%m-%d')

print(f"Max Single-Day Gain: {max_gain['Daily Return']:.2%} (Date: {date_gain})")
print(f"Max Single-Day Loss: {max_loss['Daily Return']:.2%} (Date: {date_loss})")

# Calculate overall volatility (Standard Deviation)
volatility = df['Daily Return'].std()
print(f"Average Daily Volatility (Std Dev): {volatility:.2%}")

print("-" * 30)
print("Analysis complete.")
```
